```python
# Define columns to check for outliers
column1 = 'socialNbFollowers'  # Column from ds1_df
column2 = 'totalproductsbought'  # Column from ds2_df

# Ensure the columns are numeric
ds1_df[column1] = pd.to_numeric(ds1_df[column1], errors='coerce')
ds2_df[column2] = pd.to_numeric(ds2_df[column2], errors='coerce')

# Compute IQR for ds1_df
Q1_ds1 = ds1_df[column1].quantile(0.25)
Q3_ds1 = ds1_df[column1].quantile(0.75)
IQR_ds1 = Q3_ds1 - Q1_ds1

# Define bounds for ds1_df
lower_bound_ds1 = Q1_ds1 - 1.5 * IQR_ds1
upper_bound_ds1 = Q3_ds1 + 1.5 * IQR_ds1

# Detect outliers in ds1_df
outliers_ds1 = ds1_df[(ds1_df[column1] < lower_bound_ds1) | (ds1_df[column1] > upper_bound_ds1)]

# Compute IQR for ds2_df
Q1_ds2 = ds2_df[column2].quantile(0.25)
Q3_ds2 = ds2_df[column2].quantile(0.75)
IQR_ds2 = Q3_ds2 - Q1_ds2

# Define bounds for ds2_df
lower_bound_ds2 = Q1_ds2 - 1.5 * IQR_ds2
upper_bound_ds2 = Q3_ds2 + 1.5 * IQR_ds2

# Detect outliers in ds2_df
outliers_ds2 = ds2_df[(ds2_df[column2] < lower_bound_ds2) | (ds2_df[column2] > upper_bound_ds2)]

# Print outliers info
print(f"Number of outliers in {column1} (ds1_df): {outliers_ds1.shape[0]}")
print(outliers_ds1)

print(f"Number of outliers in {column2} (ds2_df): {outliers_ds2.shape[0]}")
print(outliers_ds2)
```

```python
# Assuming '' is the common key, merge both DataFrames
unified_df = pd.merge(ds1_df, ds2_df, on='country')

# Save the combined DataFrame to a new CSV file
combined_csv_file = 'combined_data.csv'
unified_df.to_csv(combined_csv_file, index=False)

print(f"Combined data saved to {combined_csv_file}")
```
